```python
import numpy as np
```

```python
no_of_pts= 500
x_pts=[]
y_pts= []
```

```python
a=0.22
b=0.78
```

```python
for i in range(no_of_pts):
    x=np.random.normal(0.0, 0.5)
    y=a*x + b+ np.random.normal(0.0, 0.1)
    x_pts.append([x])
    y_pts.append([y])
```

```python
import matplotlib.pyplot as plt
plt.plot(x_pts, y_pts, 'o', label='Input Data')
plt.legend()
plt.show()
```

# Cost Function and Gradient Descent

##### Step 1

```python
import tensorflow.compat.v1 as tf  #use this if you have the latest version of tensorflow ie. tensorflow 2.3.0
tf.disable_v2_behavior()
```

##### Step 2

```python
#Define A and b unknowns
A= tf.Variable(tf.random.uniform([1], -1.0, 1.0))
b= tf.Variable(tf.zeros([1]))

#Instantiate the relationship
y= A*x_pts +b
```

##### Step 3

```python
#cost function
cost = tf.reduce_mean(tf.square(y-y_pts)) #taking Mean Squared Error as the Cost Function
```

##### Step 4

```python
#optimizer
optimizer= tf.train.GradientDescentOptimizer(learning_rate= 0.5)
```

##### Step 5

### Training Model

```python
train= optimizer.minimize(cost)
```

##### Step 6

### Testing Model

```python
#instantiating the variables
model= tf.global_variables_initializer()
```

```python
#setting the iteration
with tf.Session() as session:
    #simulating the model
    session.run(model)
    for step in range(0,31): #considering upto 30 steps. 
        session.run(train)
        if (step % 5)== 0:
            plt.plot(x_pts, y_pts, 'o', label='step={}'.format(step))
            plt.plot(x_pts, session.run(A)*x_pts + session.run(b), color='orange')

            plt.legend()

            plt.show()
```
